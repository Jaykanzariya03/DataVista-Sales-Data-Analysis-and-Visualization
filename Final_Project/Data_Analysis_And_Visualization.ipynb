## DATA VISUALIZATION

#### Importing Libraries

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MaxNLocator
from matplotlib.dates import DateFormatter, MonthLocator
import seaborn as sns
```

#### 1) Bar and line plot for showing yearly sales and order counts 

```python
auto_sales_data['ORDER_DATE'] = pd.to_datetime(auto_sales_data['ORDER_DATE'])

# Calculate yearly total sales
yearly_sales = auto_sales_data.groupby(pd.Grouper(key='ORDER_DATE', freq='Y'))['TOTAL_PRICE'].sum()

# Calculate yearly order count (unique ORDER_NUMBERs per year)
yearly_order_count = auto_sales_data.groupby(pd.Grouper(key='ORDER_DATE', freq='Y'))['ORDER_NUMBER'].nunique()

# Create the figure and primary axis for total sales
fig, ax1 = plt.subplots()

# Bar plot for total sales
bars = ax1.bar(yearly_sales.index.year, yearly_sales.values, width=0.6, color='skyblue', label='Total Sales')

# Annotate bars with the total sales values
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height, f'${height:,.0f}', ha='center', va='bottom')

# Format y-axis of total sales as currency
ax1.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax1.set_ylabel('Total Sales ($)', color='skyblue')
ax1.set_xlabel('Sales Year')
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))  # Ensure x-axis shows integer years

# Create secondary y-axis for order count
ax2 = ax1.twinx()

# Line plot for yearly order count
ax2.plot(yearly_order_count.index.year, yearly_order_count.values, color='orange', marker='o', linestyle='-', label='Order Count')
ax2.set_ylabel('Order Count', color='orange')

# Title and layout adjustments
plt.title('Total Sales and Order Count by Year')
fig.tight_layout()  # Adjust layout for better spacing

plt.show()
```

There was a significant increase in both total sales and order count from 2018 to 2019, reaching a peak in 2019.

#### 2) Line plot for monthly sales over time

```python
monthly_sales = auto_sales_data.groupby(pd.Grouper(key='ORDER_DATE', freq='M'))['TOTAL_PRICE'].sum()

plt.plot(monthly_sales.index, monthly_sales.values)

formatter = FuncFormatter(lambda x, _: f'${x:,.0f}')

plt.gca().yaxis.set_major_formatter(formatter)

plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))

plt.xticks(rotation=45)

plt.title('Monthly Sales')

plt.show()
```

Peaks around the end of each year suggest seasonal trends or significant events driving higher sales.

#### 3)Multiple Line plot for monthly sales by year

```python
auto_sales_data['ORDER_DATE'] = pd.to_datetime(auto_sales_data['ORDER_DATE'], errors='coerce')

# Ensure TOTAL_PRICE is numeric
auto_sales_data['TOTAL_PRICE'] = pd.to_numeric(auto_sales_data['TOTAL_PRICE'], errors='coerce')

# Extract Year and Month from ORDER_DATE
auto_sales_data['YEAR'] = auto_sales_data['ORDER_DATE'].dt.year
auto_sales_data['MONTH'] = auto_sales_data['ORDER_DATE'].dt.month

# Group by YEAR and MONTH, then calculate total sales for each month
monthly_sales_trend = auto_sales_data.groupby(['YEAR', 'MONTH'])['TOTAL_PRICE'].sum().unstack(level=0, fill_value=0)

# Plotting the line chart
plt.figure(figsize=(11, 4))

# Plot a separate line for each year
for year in monthly_sales_trend.columns:
    plt.plot(monthly_sales_trend.index, monthly_sales_trend[year], label=f'Year {year}', marker='o')

# Labeling the chart
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend by Year')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year')
plt.grid(True)

plt.ticklabel_format(axis='y', style='plain')
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('${x:,.0f}'))



plt.tight_layout()
plt.show()
```

Stable sales were recorded in 2018 with a peak in November.
Gradual sales increase in 2019, peaking again in November.


#### 4)Total Sales of Each Product Line by Year

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data (replace with your actual DataFrame)
# Make sure `auto_sales_data` is loaded as a DataFrame

# Ensure ORDER_DATE is in datetime format
auto_sales_data['ORDER_DATE'] = pd.to_datetime(auto_sales_data['ORDER_DATE'], errors='coerce')

# Ensure TOTAL_PRICE is numeric
auto_sales_data['TOTAL_PRICE'] = pd.to_numeric(auto_sales_data['TOTAL_PRICE'], errors='coerce')

# Drop rows with NaN in ORDER_DATE or TOTAL_PRICE
auto_sales_data = auto_sales_data.dropna(subset=['ORDER_DATE', 'TOTAL_PRICE'])

# Filter for data in 2018, 2019, 2020
auto_sales_data = auto_sales_data[(auto_sales_data['ORDER_DATE'].dt.year >= 2018) & 
                                  (auto_sales_data['ORDER_DATE'].dt.year <= 2020)]

# Extract YEAR from ORDER_DATE
auto_sales_data['YEAR'] = auto_sales_data['ORDER_DATE'].dt.year

# Group by PRODUCT_LINE and YEAR, then calculate total sales for each group
sales_by_product_year = auto_sales_data.groupby(['PRODUCT_LINE', 'YEAR'])['TOTAL_PRICE'].sum().unstack(fill_value=0)

# Create a numerical index for PRODUCT_LINE for plotting
product_lines = sales_by_product_year.index.tolist()
indices = range(len(product_lines))

# Set up bar width and figure size
width = 0.25
fig, ax = plt.subplots(figsize=(12, 8))

# Convert years to integer to use in bar positions
years = sales_by_product_year.columns

# Plot bars for each year
for i, year in enumerate(years):
    ax.bar([index + width * i for index in indices], 
           sales_by_product_year[year], 
           width=width, 
           label=str(year))

# Labeling the chart
ax.set_xlabel('Product Line')
ax.set_ylabel('Total Sales')
ax.set_title('Total Sales of Each Product Line by Year')
ax.set_xticks([index + width for index in indices])
ax.set_xticklabels(product_lines, rotation=45, ha='right')
ax.legend(title='Year')

plt.tight_layout()
plt.show()

```

â€Classic Carsâ€ had the highest sales in 2019, while â€Trainsâ€ had the lowest consistently.

#### 6)Quantity and average price of each product

```python
# Summing quantity ordered for each product line
quantity_each_product = auto_sales_data.groupby('PRODUCT_LINE')['QUANTITY_ORDERED'].sum()

# Calculating average price for each product line
avg_price_each_product = auto_sales_data.groupby('PRODUCT_LINE').apply(lambda x: (x['TOTAL_PRICE'].sum() / x['QUANTITY_ORDERED'].sum()))

# Create the figure and primary axis for total sales
fig, ax1 = plt.subplots()

bars = ax1.bar(quantity_each_product.index, quantity_each_product.values, width=0.6, color='skyblue', label='Quantity')

ax1.set_ylabel('Quantity', color='skyblue')
ax1.set_xlabel('Products')

plt.xticks(rotation=45, ha='right')

# Create secondary y-axis for order count
ax2 = ax1.twinx()

ax2.plot(avg_price_each_product.index, avg_price_each_product.values, color='orange', marker='o', linestyle='-', label='Order Count')
ax2.set_ylabel('Avg price each', color='orange')

plt.title('Quantity and average price of each product')

fig.tight_layout()  # Adjust layout for better spacing

plt.show()
```

Classic Cars command a high quantity and price, indicating strong demand and possibly premium positioning. Trains have the lowest quantity and price, suggesting lower demand or a different market segment.


#### 7)Total Sales by Country and City

```python
import pandas as pd
import plotly.express as px

# Grouping the data by Country and City, calculating total sales for each City
sales_by_country_city = auto_sales_data.groupby(['COUNTRY', 'CITY'], as_index=False)['TOTAL_PRICE'].sum()

# Creating the treemap plot
fig = px.treemap(
    sales_by_country_city,
    path=['COUNTRY', 'CITY'],  # Define the hierarchy for the treemap
    values='TOTAL_PRICE',       # Use total sales as the size of each rectangle
    color='TOTAL_PRICE',        # Color based on total sales
    color_continuous_scale='Blues',  # Choose a color scale
    title='Total Sales by Country and City'
)

# Display the plot
fig.show()
```

High Sales Concentration: The USA, particularly cities like San Rafael and NYC, shows the highest concentration of sales, suggesting these are key markets.
**Madrid, Spain:** A significant sales contributor, indicating a strong market presence in Spain.

```python
sales_per_productline = auto_sales_data.groupby(['ORDER_LINE_NUMBER'])['TOTAL_PRICE'].sum()

orders_per_productline = auto_sales_data.groupby(['ORDER_LINE_NUMBER'])['ORDER_NUMBER'].nunique()

fig, ax1 = plt.subplots()

bars = ax1.bar(sales_per_productline.index,sales_per_productline.values, width=0.6, color='skyblue', label='Total Sales')

# Format y-axis of total sales as currency
ax1.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax1.set_ylabel('Total Sales ($)', color='skyblue')
ax1.set_xlabel('Order line')
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))  # Ensure x-axis shows integer years

# Create secondary y-axis for order count
ax2 = ax1.twinx()

# Line plot for yearly order count
ax2.plot(orders_per_productline.index, orders_per_productline.values, color='orange', marker='o', linestyle='-', label='Order Count')
ax2.set_ylabel('Order Count', color='orange')

# Title and layout adjustments
plt.title('Sales and order count by order line')
fig.tight_layout()  # Adjust layout for better spacing

plt.show()
```

#### 9)Monthly Trends in Deal Size

```python
auto_sales_data['ORDER_DATE'] = pd.to_datetime(auto_sales_data['ORDER_DATE'])

# Extract month and year for easier grouping
auto_sales_data['YearMonth'] = auto_sales_data['ORDER_DATE'].dt.to_period('M')

# Group the data by 'DEAL_SIZE' and 'YearMonth' to get the count of orders for each combination
monthly_deal_size_counts = auto_sales_data.groupby(['DEAL_SIZE', 'YearMonth']).size().unstack(fill_value=0)

plt.figure(figsize=(13, 6))

for deal_size in monthly_deal_size_counts.index:
    plt.plot(monthly_deal_size_counts.columns.astype(str), 
             monthly_deal_size_counts.loc[deal_size], 
             label=deal_size, marker='o')

# Customize the plot
plt.title('Monthly Trends in Deal Size')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Deal Size')

plt.tight_layout()
plt.show()
```

Significant spikes in small and medium deals in October of 2018 and 2019. Large deals remained low and stable over time.
